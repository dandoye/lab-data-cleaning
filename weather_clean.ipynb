```python
import pandas as pd
import numpy as np
```

```python
weather = pd.read_csv('weather-raw.csv')
weather.head()
```

```python
weather = weather.fillna(0)
weather.head()
```

```python
drop_list = ['id','year']
weather = weather.drop(drop_list, axis = 1)
weather.head()
```

```python
weather2 = weather.melt(id_vars=['month','element'])
weather2.head()
```

```python
weather3 = weather2.pivot_table(index = ['month','variable'], columns = 'element')
weather3.head()
```

```python
weather4 = weather3.loc[(weather3!=0).any(axis=1)]
weather4.head()
```

```python
weather5 = weather4.reset_index()
weather5.columns = ("month","day","tmax","tmin")
weather5.head()
```

```python
# Adding column for weather temperature differences - assuming data set is potentially reocrding

weather5["tdiff"] = weather5["tmax"].sub(weather5["tmin"], axis = 0)
weather5.head(30)
```
